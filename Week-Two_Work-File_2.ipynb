```python
def multiply(*numbers):
    result = 1
    for number in numbers:
        result = result * number

    return result


print(multiply(190))
```

```python
def numOrProduct(num1,*numbers):
    if len(numbers) == 0:
        return num1
    else:
        result = num1
    for number in numbers:
        result = result * number

    return result
    


print(numOrProduct(190,3,4,5))
```
