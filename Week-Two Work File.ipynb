```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        print(n % i)
        if n % i == 0:
            
            return False
    return True

# print(is_prime(1014525))  # Output: False
print(is_prime(101))  # Output: True
```
